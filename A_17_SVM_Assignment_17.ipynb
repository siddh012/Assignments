```python

import pandas as pd
import numpy as np

```

```python
df_test=pd.read_csv("https://raw.githubusercontent.com/omgo101/dataset-/main/SalaryData_Test.csv")
```

```python
df_test
```

```python
df_train=pd.read_csv("https://raw.githubusercontent.com/omgo101/dataset-/main/SalaryData_Train(1).csv")
df_train
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df_test['Salary']=lb.fit_transform(df_test['Salary'])
df_train['Salary']=lb.fit_transform(df_train['Salary'])
```

```python
df1_test=pd.get_dummies(df_test)
df1_train=pd.get_dummies(df_train)
```

```python
x_train=df1_train.drop('Salary',axis=1)
y_train=df1_train['Salary']
x_test=df1_test.drop('Salary',axis=1)
y_test=df1_test['Salary']
```

```python
y_train
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

# running svm with linear kernel

```python
linear_svc=SVC(kernel='linear', C=1.0)
# fit classifier to training set
linear_svc.fit(x_train,y_train) 
pred=linear_svc.predict(x_test)
from sklearn.metrics import accuracy_score
# compute and print accuracy score
print('Model accuracy score with linear kernel and C=1.0 : {0:0.4f}'. format(accuracy_score(y_test, pred)))
```

# running SVM with polynomial kernel

```python
linear_svc=SVC(kernel='poly', C=1.0)
# fit classifier to training set
linear_svc.fit(x_train,y_train) 
pred=linear_svc.predict(x_test)
from sklearn.metrics import accuracy_score
# compute and print accuracy score
print('Model accuracy score with polynomial kernel and C=1.0 : {0:0.4f}'. format(accuracy_score(y_test, pred)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(pred,y_test))
print(classification_report(pred,y_test))
```

```python
np.mean(pred==y_test)*100
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
forest['y']=lb.fit_transform(forest['size_category'])
```

```python
forest
```

```python
forest=forest.drop(['month','day','size_category'],axis=1)
```

```python
forest
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=forest.drop('y',axis=1)
y=forest['y']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=101)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf','linear', 'poly']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid_predictions=grid.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,grid_predictions))
print('\n')
print(classification_report(y_test,grid_predictions))
```
